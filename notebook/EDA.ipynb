```python
#carregando as bibliotecas
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("../datasets/vehicles.csv") #carregando o dataset
df.head()
```

```python

df.info() #mostra informações sobre o dataset
```

```python
df['price']=df['price'].astype(float) # convertendo a coluna price para float
df['model_year']=df['model_year'].fillna(df['model_year'].mean()) # preenchendo valores nulos na coluna model_year com a média
df['model_year']=df['model_year'].astype(int) # convertendo a coluna model_year para int
df['cylinders'] = df['cylinders'].fillna(df.groupby('model')['cylinders'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan))# substituindo valores nulos na coluna cylinders pela moda do respectivo modelo
df['cylinders'] = df['cylinders'].astype(int) # convertendo a coluna cylinders para int
df['odometer'] = df['odometer'].fillna(df.groupby('model_year')['odometer'].transform('mean')) # substituindo valores nulos na coluna odometer pela média do respectivo ano do modelo
df['paint_color'] = df['paint_color'].fillna('unknown') # substituindo valores nulos na coluna paint_color por 'unknown'
df['is_4wd'] = df['is_4wd'].fillna(0).astype(int) # substituindo valores nulos na coluna is_4wd por 0
df['date_posted'] = pd.to_datetime(df['date_posted']) # convertendo a coluna date_posted para datetime
df['month_posted'] = df['date_posted'].dt.month # criando a coluna month_posted
```

```python
df_mean_year = df.groupby('model_year')['price'].mean().reset_index(name='mean_price')

fig = px.bar(df_mean_year, x='model_year',
             y='mean_price',
             title='Preço médio por ano do modelo',
             labels={'model_year': 'Ano do Modelo', 'mean_price': 'Preço Médio'})
fig.show()

```

```python
fig = px.histogram(df, x= 'model_year',
                   nbins=30, title='Distribuição dos anos dos modelos dos veículos',
                   labels={'model_year': 'Ano do Modelo', 'count': 'Contagem'})
fig.show()
```

```python
df_mean_type = df.groupby('type')['price'].mean().reset_index(name='mean_price')
fig = px.bar(df_mean_type, x='type',
             y='mean_price',
             title='Preço médio por tipo de veículo',
             labels={'type': 'Tipo de Veículo', 'mean_price': 'Preço Médio'})   
fig.show()
```
